```python
# .loc - selects a label -- for example a SKU ID like "SKU-003" or a date like "2026-03-14"

# .iloc -- selects by integer position -- using the row number similar to an index or python list

# import pandas
import pandas as pd
```

```python
import pandas as pd

# create the dataset

df = (
    pd.DataFrame(
    {
        "sku_id" : ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
        "category" : ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
        "warehouse" : ["East", "West", "East", "East", "West", "West"],
        "units_on_hand" : [120, 45, 300, 0, 88, 210],
        "unit_cost" : [25.50, 42.00, 15.75, 15.75, 8.25, 8.25]
    }
    )
.set_index("sku_id")
)
print(df)
```

```python
# exercise 2

# select the entire row for "SKU-003"
# df.loc[row label]

df.loc["SKU-003"]
print(df.loc["SKU-003"])

# select the unit_cost value for "SKU-003"
# .loc[row label, column label]

df.loc["SKU-003", "unit_cost"]
print(df.loc["SKU-003", "unit_cost"])

# select rows "SKU-002" through "SKU-005"
# .loc[]'s end label is inclusive which differs from python lists as [1:3] only returns 1 and 2
df.loc["SKU-002":"SKU-005"]
print(df.loc["SKU-002":"SKU-005"])
```

```python
# Exercise 3 
# .iloc[] by position

# select the first row using .iloc[]
df.iloc[0]
print(df.iloc[0])

# select the last row using .iloc[]
df.iloc[-1]
print(df.iloc[-1])

# select rows 1 through 3 (2nd, 3rd, and 4th) using .iloc[]
# based on results .iloc does not appear to be inclusive
df.iloc[1:4]
print(df.iloc[1:4])
```

```python
# exercise 4

# use .loc with a list of labels
# select units on hand for "SKU-001" & "SKU-004"

df.loc[["SKU-001", "SKU-004"], "units_on_hand"]

# select the same two rows but use their integer positions
# what are the row-positions?
# what are the column-positions?
# when you set a column as an index it is no longer a column

# verify your columns
print(df.columns.tolist())

# "units_on_hand" is in position 2
df.iloc[[0,3], 2]

# when would you use .iloc[] over .loc[]
# .iloc[] is used to slice data by their integer positions
# whereas .loc[ ] would be used by a supply chain data scientist to query SKU's, supplier names, geographies, and etc.
```
