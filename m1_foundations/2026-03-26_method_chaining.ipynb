```python
# recreate the inventory dataframe
# do not set sku_id as the index -- keep it as a regular column
import pandas as pd

# create the dataset as a flat dataframe

df = (
    pd.DataFrame(
    {
        "sku_id" : ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
        "category" : ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
        "warehouse" : ["East", "West", "East", "East", "West", "West"],
        "units_on_hand" : [120, 45, 300, 0, 88, 210],
        "unit_cost" : [25.50, 42.00, 15.75, 15.75, 8.25, 8.25]
    }
    )
    )

print(df)
```

```python
# exercise 2
# write traditional "non-chained" pandas w/intermediate variables
trad1 = df[df["units_on_hand"] > 0]
trad2 = trad1.assign(inventory_value = trad1["units_on_hand"] * trad1["unit_cost"])
trad3 = trad2.sort_values(by=["inventory_value"], ascending=False)
print(trad3.loc[:, ["sku_id", "category", "units_on_hand", "inventory_value"]])
```

```python
# write exercise 2 with method chaining
chain = (
    df
    .query("units_on_hand > 0")
    .assign(inventory_value=lambda x: x["units_on_hand"] * x["unit_cost"])
    .sort_values(by=["inventory_value"], ascending=False)
    .loc[:, ["sku_id", "category", "units_on_hand", "inventory_value"]]
)

print(chain)
```
