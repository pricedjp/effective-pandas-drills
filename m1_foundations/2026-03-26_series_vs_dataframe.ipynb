```python

#The mental model before you write any code:**

#A **Series** is a single column of data with an index. 
# Think of it as a one-dimensional labeled array — or column - every value has a position (the index) and a value.

# A **DataFrame** is a collection of Series that share the same index. 
# Think of it as a dictionary where every key is a column name and every value is a Series. That's literally how it's built internally.
```

```python
# exercise 1

import pandas as pd
import numpy as np

units = pd.Series([120, 45, 300, 0, 88, 210])
print(units)

# what is its data type?
print(units.dtype)

# print the index
print(units.index)

# print its values
print(units.values)
```

```python
# exercise 2

# create the same series but pass an index argument using these SKU ID's
# SKU IDs: ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"]
# then access "SKU-003" using the index label

units = pd.Series(data=[120, 45, 300, 0, 88, 210],
                  index=["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"]
                  )

print(units["SKU-003"])
```

```python
# exercise 3
# create two series: units and cost
# use the same sku index
# then combine them into a dataframe
import pandas as pd
import numpy as np

units = pd.Series(data=[120, 45, 300, 0, 88, 210],
                  index=["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"]
                )

cost = pd.Series(data=[25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
                 index=["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"]
                 )

# use dictionary syntax
# make a column called units using units series
df = pd.DataFrame({"units":units, 
                  "cost":cost}
                  )
print(df)

# columns
print(df.columns)

# index
print(df.index)

# df["units"]
print(df["units"])
print(type(df["units"]))
```
