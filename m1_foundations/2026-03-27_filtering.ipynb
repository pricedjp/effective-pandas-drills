```python
#import libraries

import pandas as pd
import numpy as np

# create the dataset as a flat dataframe

df = (
    pd.DataFrame(
    {
        "sku_id" : ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
        "category" : ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
        "warehouse" : ["East", "West", "East", "East", "West", "West"],
        "units_on_hand" : [120, 45, 300, 0, 88, 210],
        "unit_cost" : [25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
        "reorder_point" : [50, 30, 100, 100, 40, 40]
    }
    )
    )

print(df)
```

```python
# exercise 2 Boolean Masks
# & = And
# | = OR

# all sku's where units_on_hand is > 50
mask1 = df[df["units_on_hand"] > 50]
print(mask1)

# all sku's where category is "Widget" an units_on_hand is > 0
mask2 = df[(df["category"]=="Widget") & (df["units_on_hand"] > 0)]
print(mask2)

# all sku's where unit_cost is < 10 OR units_on_hand > 200
mask3 = df[(df["unit_cost"] < 10) | (df["units_on_hand"] > 200)]
print(mask3)
```

```python
# exercise 3
# rewrite all 3 filters using .query()
# note: the conditions inside .query() must be a string
# use "and" and "or" when using .query()

# all sku's where units_on_hand is > 50
mask4 = (
    df
    .query("units_on_hand > 50")
)
print(mask4)

# all sku's where category is "Widget" an units_on_hand is > 0
mask5 = (
    df
    .query("category == 'Widget' and units_on_hand > 0")
)
print(mask5)
# all sku's where unit_cost is < 10 OR units_on_hand > 200
mask6 = (
    df
    .query("unit_cost < 10 or units_on_hand > 200")
)
print(mask6)


```

```python
# exercise 4
# use .loc[row, col] with a boolean mask

# all sku's where units_on_hand is > 50
mask7 = df.loc[
    df["units_on_hand"] > 50, ]
print(mask7)

# all sku's where category is "Widget" an units_on_hand is > 0
mask8 = df.loc[
    (df["category"]=="Widget") 
    & 
    (df["units_on_hand"]> 0),
]
print(mask8)

# all sku's where unit_cost is < 10 OR units_on_hand > 200
mask9 = df.loc[
    (df["unit_cost"] < 10)
    |
    (df["units_on_hand"] > 200),
]
print(mask9)

# all sku's where units_on_hand < reorder_point
# only show sku_id, category, units_on_hand, and reorder_point
mask10 = df.loc[
    df["units_on_hand"] < df["reorder_point"],
    ["sku_id", "category", "units_on_hand", "reorder_point"]
]
print(mask10)
```

```python
# exercise 5

# When would you use a boolean mask over .query()?
# used if you need to store the mask as a variable and reuse it -- or the condition includes a py variable
# both boolean masks and .query() return all columns
# inside a boolean mask every column reference needs a wrapper

# When would you use .query() over a boolean mask?
# use .query() over a boolean mask inside method chains -- it is easier to read and avoids double df[df[]] wrappers
# .query() uses "and" and "or" not "&" and "|" -- plain english is acceptable
# .query() passes the condition as text
# this makes .query() cleaner to read
# .query() is a method called on a dataframe 
# you do not need to use the df[] outside and df[] inside wrappers
# string values still need quotes with "=="

# When would you use .loc[] over the other two?
# use .loc[] when you need to filter rows AND also select specific columns
# you cannot also select columns from .query() -- in .loc[] this an be done in the same expression [row,col]
# the outside df[] wrapper says filter this dataframe
# the inside df[] wrapper says using this column
```
