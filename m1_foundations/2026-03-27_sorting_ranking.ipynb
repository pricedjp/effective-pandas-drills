```python
# module 1.5
# sort_values() -- reorders rows by the values in one or more columns

# sort_index() -- reorders rows by the index label instead of the column name

# .rank() -- no reordering. creates a new column that shows where a row falls in comparison to others (1st or 5th etc). maintains original column order.
```

```python
# exercise 1

# create the data frame and import libraries
import pandas as pd

# df
# add a monthly_demand column

df = (
    pd.DataFrame(
    {
        "sku_id" : ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
        "category" : ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
        "warehouse" : ["East", "West", "East", "East", "West", "West"],
        "units_on_hand" : [120, 45, 300, 0, 88, 210],
        "unit_cost" : [25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
        "reorder_point" : [50, 30, 100, 100, 40, 40],
        "monthly_demand" : [80, 60, 120, 90, 70, 95]
    }
    )
    )
print(df)
```

```python
# sort by units_on_hand ascending
s1 = df.sort_values(by="units_on_hand",
                    ascending=True)
print(s1)
```

```python
# sort by unit_cost descending
s2 = df.sort_values(by="unit_cost",
                    ascending=False)
print(s2)
```

```python
# sort by category ascending
# then sort by units_on_hand descending
s3 = df.sort_values(by=["category", "units_on_hand"],
                    ascending=[True, False]
                    )
print(s3)
```

```python
# sort by units_on_hand descending
# place .NaN values last using na_position
s4 = df.sort_values(by="units_on_hand",
                    ascending=False,
                    na_position="last")
print(s4)
```

```python
# add a demand_rank column 
# this column will compare the monthly_demand values found in each row ultimately ranking sku's by monthly_demand
# high to low so ascending=False
# highest demand = 1

# .rank()
# returns float values by default
# the "average" method parameter averages tied positions
# if there are no ties then .rank()returns whole numbers

# method parameters for ties
# in supply chain work -- rank often represents a tier not a competition so "dense" is best
# average — tied items share the average of their positions (default, returns floats)
# min — tied items all get the lowest position number
# max — tied items all get the highest position number
# first — tied items ranked in the order they appear
# dense — like min but no gaps in ranking numbers -- best for supply chain work

df = (
    df
    .assign(
        demand_rank=lambda x:x["monthly_demand"].rank(ascending=False, method="dense")
    )
)

print(df[["sku_id", "monthly_demand", "demand_rank"]])
```
