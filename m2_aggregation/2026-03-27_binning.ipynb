```python
# binning converts continuous numeric values into discrete categories
# pd.cut() -- equal width bins where you define the bin edges
# example if the range is (0, 5000) then the bins are 0-1667 since 5000 / 3 = 1667, then 1667-3333, then 3333-5000
# pd.qcut() -- quantile bins where you define the # of bins. each bin contains the same # of rows.

# create the data
# import python libraries

import pandas as pd
import numpy as np

# data
# add inventory_value using .assign() from previous lesson

df = (
    pd.DataFrame(
    {
        "sku_id" : ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
        "category" : ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
        "warehouse" : ["East", "West", "East", "East", "West", "West"],
        "units_on_hand" : [120, 45, 300, 0, 88, 210],
        "unit_cost" : [25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
        "reorder_point" : [50, 30, 100, 100, 40, 40],
        "monthly_demand" : [80, 60, 120, 90, 70, 95],
    })
    .assign(inventory_value=lambda x: x["units_on_hand"] * x["unit_cost"])
)

print(df)

```

```python
# exercise 2
# use pd.cut() w/custom bin-edges. left-edge is exclusive, the right-edge is inclusive
# (0-> Exclusive, 500->Inclusive] in this example 0 will fall outside the first -bin so you have to include_lowest=True
# add an abc_tier column
# use the chain method
# C = 0-500
# B = 500 to 3000
# A = > 3000
# question: are the bins inclusive?
df = (
    df
    .assign(
        abc_tier=pd.cut(
            df["inventory_value"],
            bins=[0, 500, 3000, float("inf")],
            labels=["C", "B", "A"],
            include_lowest=True
        )
    )
)

print(df[["sku_id", "inventory_value", "abc_tier"]])
```

```python
# exercise 3

# get an idea of the range of values for the 3 bins
# look at the monthly_demand values
df["monthly_demand"].sort_values(ascending=False)

# use pd.qcut() on monthly_demand w/3 equal-frequncy bins
# categorize and label each bin

# use pd.cut() when you need to control the bin sizes and the boundaries are important --> Classification
# use pd.qcut() when you need to see an equal frequency across a column of data --> Balanced groupings of supplier performance

df = (
    df
    .assign(
        demand_tier=pd.qcut(
            df["monthly_demand"],
            q=3,
            labels=["low", "mid", "high"]
        )
    )
)

print(df[["sku_id", "monthly_demand", "demand_tier"]])
```
