```python
# create the data
# import python libraries

import pandas as pd
import numpy as np

# data
# df 

df = (
    pd.DataFrame(
    {
        "sku_id" : ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
        "category" : ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
        "warehouse" : ["East", "West", "East", "East", "West", "West"],
        "units_on_hand" : [120, 45, 300, 0, 88, 210],
        "unit_cost" : [25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
        "reorder_point" : [50, 30, 100, 100, 40, 40],
        "monthly_demand" : [80, 60, 120, 90, 70, 95],
    })
)

print(df)
```

```python
# exercise 2
# with .agg(
# output_name = ("source_column", "function")
# )
# what becomes the index? the by=category sets the index
gb = (df
 .groupby(
     by="warehouse"
     )
 .agg(
     total_units=("units_on_hand", "sum"),
     avg_cost=("unit_cost", "mean"),
     sku_count=("sku_id", "count")
     )
     )

print(gb)
```

```python
# exercise 3
# .groupby(by="category")["units_on_hand"]
# splits the data into the 3 categories --> widget, gadget, and component
# then selects the units_on_hand column for each group
# .transform("sum") calculates the sum for each group then broadcasts it back to every row in that group (120 + 45 = 165)
# .transform() returns the same shape as the original dataframe
# sum the units_on_hand values down the column by category --> place them into category_total_units
df = (
    df
    .assign(
        category_total_units=lambda x: x.groupby(by="category")["units_on_hand"].transform("sum"),
        pct_of_category=lambda x: (x["units_on_hand"] / x["category_total_units"] * 100).round(1)
    )
)

print(df[["sku_id", "category", "units_on_hand", "category_total_units", "pct_of_category"]])
```

```python
# exercise 4
# use .filter() to keep only categories where total units_on_hand > 200
# this filters by the category_total_units value
filtered_df = (
    df
    .groupby(by="category")
    .filter(lambda x: x["units_on_hand"].sum() > 200)
)

print(filtered_df[["sku_id", "category", "units_on_hand"]])
```
