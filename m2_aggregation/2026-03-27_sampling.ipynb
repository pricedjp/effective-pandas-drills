```python
# M2.5 -- Sampling
# Random Sampling -- grab N random rows
# Fractional Sampling -- grab a percentage of rows
# Stratified Sampling -- grab a proportional sample from each group so every category is represented

# there is a random_state parameter that must be used for reproducability
# you set a seed so you get the same random sample every time you run the code
# this is important for auditing and reporting that another team member may use to reproduce your results
```

```python
# import
import pandas as pd
import numpy as np

# load sc inventory data with a larger number of SKU's
# f"SKU-{str(i).zfill(3)}" generates sequential SKU ID's
# .zfill() pads the number with leading zeros to the length of X number of character
# .str(i) goes from 1,2,...10 etc.
# .zfill(3) pads to the length of 3 characters -- making the string at least 3 characters wide "001" up to "999"
# whereas .zfill(2) would make the string at least 2 characters wide "01"
df = pd.DataFrame({
    "sku_id": [f"SKU-{str(i).zfill(3)}" for i in range(1,13)],
    "category": ["Widget"]*4 + ["Gadget"]*4 + ["Component"]*4,
    "warehouse": ["East", "West"] * 6,
    "units_on_hand": [120, 45, 300, 0, 88, 210, 55, 175, 400, 30, 90, 260],
    "unit_cost": [25.50, 42.00, 15.75, 15.75, 8.25, 8.25,
                  32.00, 18.50, 5.10, 5.10, 22.75, 22.75],
    "monthly_demand": [80, 60, 120, 90, 70, 95, 85, 110, 200, 150, 65, 130]
})
print(df)
```

```python
# Sample exactly 4 rows — n=4, random_state=42
samexact = (
    df
    .sample(n=4, 
            random_state=42)
)

# Sample 30% of rows — frac=0.3, random_state=42
samfrac = (
    df
    .sample(frac=0.3,
            random_state=42)
)
# Run the same sample twice without random_state — confirm different rows each time
samnorand = (
    df
    .sample(n=4)
)

print(samexact)
print(samfrac)
print(samnorand)
```

```python
# exercise 3 -- stratified sampling
# often used in procurement due to potential audits
# ensures every category is represented
# n=1 means "1 SKU from each category" will be included in the sample

stratified = (
    df
    .groupby(by="category")
    .sample(n=1,
            random_state=42)
)

print(stratified[["sku_id", "category", "units_on_hand"]])
```

```python
# exercise 4: random_state
# what does it do? it ensures the same random sample is used every time the dataframe is sampled
# for supply chain purposes you can imagine the importance of including every supplier not just certain types
# the specific number is just a reference for the random sample you want to repeaedly use
# what matters is that you repeatedly use the same number so your sample is reproducible
```
