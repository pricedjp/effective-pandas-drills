```python
# M3.3 -- Indexing and Reindexing
# .set_index() promotes a column to become the row labels
# this column disappears from the DataFrame and becomes the index

# .reset_index() -- demotes the index back to a regular column in the DataFrame
# the index is replaced with the default 0 based integer index

# .reindex() -- conforms the DataFrame to a new set of index labels
# rows that exist are kept
# rows that are missing are filled with NaN
# this is used with time-series data where gaps in dates must be filled

# import libraries
import pandas as pd
import numpy as np

# create dataframe
df = pd.DataFrame({
    "sku_id":        ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
    "category":      ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
    "warehouse":     ["East", "West", "East", "East", "West", "West"],
    "units_on_hand": [120, 45, 300, 0, 88, 210],
    "unit_cost":     [25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
    "reorder_point": [50, 30, 100, 100, 40, 40],
    "monthly_demand":[80, 60, 120, 90, 70, 95],
    "quality_flag":  ["pass", "pass", "fail", "fail", "pass", "pass"]
})
print(df)
```

```python
# Exercise 1

# Write three operations:

# Set sku_id as the index — show the result
df_set = df.set_index(keys="sku_id")
print(df_set)
# Then reset it back to a regular column — show the result
df_reset = df.reset_index()
print(df_reset)
# Set sku_id as the index but use drop=False to keep it as a column too — show the result and add a comment explaining when this would be useful
df_keep = df.set_index(keys="sku_id", drop=False)
print(df_keep)
```

```python
# exercise 3
# set category and warehous as multi-level indexes
# this is a two-level index
# the top level label is the first label...in this case it is category
df_multi = df.set_index(["category", "warehouse"])
print(df_multi)

# select all East warehouse rows from the Gadget category
# add a tuple to .loc using ()
# returns only rhe combination Gadget + East rows
print(df_multi.loc[("Gadget", "East")])
```

```python
# exercise 4
# use .reindex() for gap filling

# create a weekly demand series
demand = pd.Series(
    [100, 150, 130, 180],
    index=pd.to_datetime(["2024-01-01", "2024-01-08", "2024-01-22", "2024-02-05"])
)
print(demand)

# create a complex weekly date range without gaps
full_index = pd.date_range("2024-01-01", "2024-02-05", freq="W-MON")
print(full_index)

# reindex to expose the gaps
demand_reindexed = demand.reindex(full_index)
print(demand_reindexed)

# fill the NaN gaps down using .ffill()
demand_filled = demand_reindexed.ffill()
print(demand_filled)

# reindex is important because it allows you to easily fill gaps in time-series data
# data can include missing weeks and data reliability can vary by vendorgit
```
