```python
# M3.2 -- .where() and .mask() and .replace()

# .where(condition) -- keeps values where the condition is True
# replaces NaN (or a specific value) where its false
# "keep this, and replace everything else"
# KEEP where True, REPLACE where False

# .mask(condition) -- the opposite of .where()
# replaces values where the condition is True
# keeps everything where it is False
# "replace this, keep everything else"
# REPLACE where True, KEEP where FALSE

# .replace() -- swaps specific values for other values
# no conditions required
# simply substitutes one value for another: a value-to-value substitution

```

```python
# import libraries
import pandas as pd
import numpy as np

# same sc data
# add a quality_flag
df = pd.DataFrame({
    "sku_id":        ["SKU-001", "SKU-002", "SKU-003", "SKU-004", "SKU-005", "SKU-006"],
    "category":      ["Widget", "Widget", "Gadget", "Gadget", "Component", "Component"],
    "warehouse":     ["East", "West", "East", "East", "West", "West"],
    "units_on_hand": [120, 45, 300, 0, 88, 210],
    "unit_cost":     [25.50, 42.00, 15.75, 15.75, 8.25, 8.25],
    "reorder_point": [50, 30, 100, 100, 40, 40],
    "monthly_demand":[80, 60, 120, 90, 70, 95],
    "quality_flag":  ["pass", "pass", "fail", "fail", "pass", "pass"]
})

print(df)
```

```python
# exercise 2
# .where()
# KEEP where True, REPLACE where FALSE

# Two operations using .assign():
# Keep units_on_hand where > 50, replace everything else with 0
# Keep unit_cost where quality_flag == "pass", replace failed SKUs with NaN

df = (
    df
    .assign(units_where = lambda x: x["units_on_hand"].where(x["units_on_hand"] > 50, 0),
            cost_where = lambda x: x["unit_cost"].where(x["quality_flag"]=="pass", np.nan))
)
print(df[["sku_id", "units_on_hand", "units_where", "unit_cost", "quality_flag", "cost_where"]])


```

```python
# exercise 3
# .mask()
# REPLACE where True, KEEP where FALSE

# Mask units_on_hand where units are 0 — replace zero-stock with NaN
# Mask unit_cost where quality_flag == "fail" — replace failed costs with NaN

df = (
    df
    .assign(
        units_on_hand_mask= lambda x: x["units_on_hand"].mask(x["units_on_hand"]==0, np.nan),
        unit_cost_mask = lambda x: x["unit_cost"].mask(x["quality_flag"]=="fail", np.nan)
    )
)
print(df)
```

```python
# use .replace()
# .replace() -- swaps specific values for other values
# .replace() is best used for simple value to value replacements that do not rely on conditional logic
df = (
    df
    .assign(
        quality_flag_clean=lambda x: x["quality_flag"].replace({"pass":"approved", "fail":"rejected"})
    )
)
print(df[["quality_flag", "quality_flag_clean"]])
```
