```python
# cumulative operations
# run a calculation that grows with each row
# each value includes all previous values
# rolling windows look back N rows but cumulative operations look back at everything from row 1 to the current row

# .cumsum() -- generates a running total -- across each row what is the sum up to and including this one?

# cummax() -- generates a running maximum -- across each row what is the highest value seen so far?

# .cumprod() generates a running product -- across each row what is the product of all rows up until this one?
```

```python
# import libraries
import pandas as pd
import numpy as np

# monthly po spend data w/suppliers

df = pd.DataFrame({
    "month": pd.date_range("2024-01-01", periods=6, freq="ME"),
    "supplier": ["Acme", "Acme", "Acme", "GlobalCo", "GlobalCo", "GlobalCo"],
    "spend":    [1500, 2200, 1800, 3100, 2800, 3400],
    "on_time_deliveries": [8, 9, 7, 10, 8, 11],
    "total_deliveries":   [10, 10, 10, 12, 11, 12]
})

print(df)
```

```python
# exercise 2
# Add two cumulative sum columns:

# cumulative_spend — running total of spend across all rows
# ytd_deliveries — running total of on_time_deliveries

# Use .assign() with lambdas. Then filter rows where cumulative_spend exceeds 5000.

df = (
    df
    .assign(
        cumulative_spend = lambda x: x["spend"].cumsum(),
        ytd_deliveries = lambda x: x["total_deliveries"].cumsum()
    )
)

print(df[df["cumulative_spend"] > 5000])
```

```python
# exercise 3
# .cummax() to identify peak spend

# .cummax() can be useful for procurement analytics 
# procurement often requires extended negotiations and knowledge of buget and spend
# specifically, procurement managers may want to know when a supplier has exceeded their spend threshold -- breaking a contract

df = (
    df
    .assign(
        peak_spend = lambda x: x["spend"].cummax()
    )
)
print(df[["month", "supplier", "spend", "peak_spend"]])
```

```python
# exercise 4
# use groupby and .cumsum() to ensure each suppliers running total resets automatically
# so if the dataframe is sorted by supplier, ex: ACME --> GlobalCo
# the groupby function will sum each of acmes spend then stop and reset when the row reaches/includes GlobalCo

df = (
    df
    .assign(
        supplier_cumspend = lambda x: x.groupby(by="supplier")["spend"].cumsum()
    )
)
# print df
print(df[["month", "supplier", "spend", "supplier_cumspend"]])

# filter for cumulative spend > 4000
# print(df[df["supplier_cumspend"] > 4000])
```
