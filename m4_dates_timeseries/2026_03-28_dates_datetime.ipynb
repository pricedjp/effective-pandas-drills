```python
# m4.1 date and time manipulation
# dates must be of the right type
# you need to convert dates to datetime64 using pd.to_datetime()
# the .dt() accessor unlocks date operations similar to the .str() accessor
```

```python
# import
import pandas as pd
import numpy as np

# create a purchase order dataset
df = pd.DataFrame({
    "po_id":       ["PO-001", "PO-002", "PO-003", "PO-004", "PO-005", "PO-006"],
    "supplier":    ["Acme", "Acme", "GlobalCo", "GlobalCo", "FastParts", "FastParts"],
    "order_date":  ["2024-01-05", "2024-02-12", "2024-01-28", 
                    "2024-03-03", "2024-02-20", "2024-03-15"],
    "receive_date":["2024-01-19", "2024-03-01", "2024-02-10",
                    "2024-03-18", "2024-03-05", "2024-04-02"],
    "amount":      [1500, 3200, 800, 4100, 650, 2900]
})
print(df)
```

```python
#Exercise 2 — Converting to datetime

#Check the dtype of order_date before conversion — it will show object (string)
# returns a string
df[["order_date"]].dtypes

#Convert both order_date and receive_date to datetime using pd.to_datetime()
# each needs their own .assign()
df_conv = (
    df
    .assign(
        order_date = lambda x: pd.to_datetime(x["order_date"]),
        receive_date = lambda x: pd.to_datetime(x["receive_date"])
        )
)
print(df_conv)

# Check the dtype again 
# now shows datetime64
print(df_conv.dtypes)
```

```python
# extract key date components from the datetime columns using .assign()
# note: .dt.dayofweek returns 0 for Monday through 6 for Sunday
# so order_dayofweek 4 = Friday
df_conv = (
    df_conv
    .assign(
        order_month=lambda x: x["order_date"].dt.month,
        order_quarter=lambda x: x["order_date"].dt.quarter,
        order_dayofweek=lambda x: x["order_date"].dt.dayofweek,
        lead_time_days=lambda x: (x["receive_date"] - x["order_date"]).dt.days
        )
)

print(df_conv[["po_id", "order_date", "order_month", "order_quarter", "order_dayofweek", "lead_time_days"]])
```

```python
# exercise 4

# All POs ordered in February 2024
po_2024 = df_conv[df_conv["order_date"].dt.month == 2]
    
print(po_2024)
# All POs with a lead time greater than 20 days
# lead_time_days is counted in days no need to convert to days -- compare directly
po_lead_gt_20 = df_conv[df_conv["lead_time_days"] > 20]
print(po_lead_gt_20[["po_id", "supplier", "lead_time_days"]])
```
