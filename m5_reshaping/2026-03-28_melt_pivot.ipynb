```python
# M5: Melt and Pivot Data
# Data comes in two shapes: (1) Wide and (2) Long
# Wide: one row per entity, many columns for each time-period or category
# Long: one row per observation, columns for the entity, period, and value

# Melt .melt()
# converts long to wide

#import libraries
import pandas as pd
import numpy as np

# create a wide format demand dataframe
df_wide = pd.DataFrame({
    "sku_id":   ["SKU-001", "SKU-002", "SKU-003"],
    "category": ["Widget", "Gadget", "Component"],
    "Jan":      [100, 60, 200],
    "Feb":      [85, 75, 180],
    "Mar":      [120, 90, 220],
    "Apr":      [95, 80, 195]
})

print(df_wide)
```

```python
# exercise 2: use .melt() to go from wide to long

# id_vars — columns to keep as identifiers (not melted)
# value_vars — columns to melt into rows
# var_name — name for the new column holding the old column names
# value_name — name for the new column holding the values

# sort by two values
# in practice you would convert "month" to datetime format
# then you would sort month second -- it would result in chronological dates ordered by sku (1st)

df_long = (
    df_wide.melt(
        id_vars=["sku_id", "category"],
        value_vars=["Jan", "Feb", "Mar", "Apr"],
        var_name="month",
        value_name="demand"
    )
    .sort_values(by=["sku_id", "month"])
)
print(df_long)

# filter for demand > 100
print(df_long[df_long["demand"] > 100])
```

```python
# exercise 3 -- using .pivot() to go from wide to tabular
df_back_to_wide = (
    df_long
    .pivot(
        index="sku_id",
        columns="month",
        values="demand"
    )
)

print(df_back_to_wide)
```
