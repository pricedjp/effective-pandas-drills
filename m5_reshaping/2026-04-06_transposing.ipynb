```python
# M5.3 Transposing Data

# Transposing flips a dataframe
# Rows become columns
# Columns become rows
# the .T attribute and .transpose() method complete the same task

# Transposing is not for reshaping analysis pipelines like .melt(), .pivot(), and .unstack()
# Transposing is for display and reporting situations where you want to flip the summary table

# import libraries
import pandas as pd
import numpy as np

# create a dataframe
# set the index

summary = pd.DataFrame({
    "Jan": [100, 60, 200],
    "Feb": [85, 75, 180],
    "Mar": [120, 90, 220],
    "Apr": [95, 80, 195]
},
    index=["Widget", "Gadget", "Component"]
)
summary.index.name = "category"
print(summary)
```

```python
# exercise 2

# Transpose the summary table two ways and confirm each is identical

# what will become the rows and what will become the columns?
# the catefories are rows that will become columns
# the months are columns that will become rows
df_T = summary.T
print(df_T)

# using .transpose()
df_transpose = summary.transpose()
print(df_transpose)

# confirm that .T == .transpose()
print(df_T.equals(df_transpose))
```

```python
# exercise 3 -- filtering on transposed data

# Filter months where Widget demand exceeded 90
print(df_T[df_T["Widget"] > 90])

# Filter months where Component demand exceeded 195
print(df_T[df_T["Component"] > 195])
```

```python
# exercise 4 -- when to use transpose
scorecard = pd.DataFrame({
    "Acme":     [95, 18, 4.2, 98],
    "GlobalCo": [88, 22, 3.8, 95],
    "FastParts":[91, 15, 4.5, 99]
}, 
index=["on_time_pct", "avg_lead_days", "quality_score", "fill_rate_pct"]
)

# note that the transposed format does a better job conveying the data in a easy to read meeting friendly style
print(scorecard)
print(scorecard.T)
```
