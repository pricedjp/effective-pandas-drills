```python
# Crosstabs
# pd.crosstab() counts the frequency of combinations between two or more categorical variables
# it is similar to a pivot-table set to count values
# how many times does a combination appear? eg East + Widget or West + Component?
# pd.Crosstab() can also normalize to 

# import libraries
import pandas as pd
import numpy as np

# create the inventory dataset
# use an f-string to create unique sku_id's
# range(1,13) generates numers 1 through 12
# str(i) converts each number to a string
# .zfill(3) pads to 3 charcters "001"
# f"SKU-{...}" adds the SKU prefix for SKU-001...SKU-012
# note: ["Widget"]*4 = ["Widget", "Widget", "Widget", "Widget"] --> * on a list repeats the list N times
# note: + on a list concatenates them end to end 
# ["Wiidget"]*2 + ["Gadget"]*2 = ["Widget", "Widget", "Gadget", "Gadget"]
df = pd.DataFrame({
    "sku_id":    [f"SKU-{str(i).zfill(3)}" for i in range(1, 13)],
    "category":  ["Widget"]*4 + ["Gadget"]*4 + ["Component"]*4,
    "warehouse": ["East", "West"] * 6,
    "supplier":  ["Acme", "Acme", "GlobalCo", "GlobalCo",
                  "Acme", "FastParts", "GlobalCo", "FastParts",
                  "Acme", "FastParts", "GlobalCo", "FastParts"],
    "units_on_hand": [120, 45, 300, 0, 88, 210, 55, 175, 400, 30, 90, 260],
    "unit_cost": [25.50, 42.00, 15.75, 15.75, 8.25, 8.25,
                  32.00, 18.50, 5.10, 5.10, 22.75, 22.75]
})

print(df)
```

```python
# exercise 2
# function signature
# pd.crosstab(
#    index,
#    columns,
#    values=None,
#    rownames=None,
#    colnames=None,
#    aggfunc=None, # sum, mean, count, min, max
#    margins=False,
#    margins_name="All",
#    dropna=True,
#    normalize=False
#)

# basic crosstab counting of SKUs by Warehouse and Category
# add margins = True
# output shows 2SKU's per warehouse category combination
# each cell = 2/12 = 0.1667
print(pd.crosstab(
    index=df["warehouse"], 
    columns=df["category"], 
)
)

# add margins
print(pd.crosstab(
    index=df["warehouse"], 
    columns=df["category"], 
    normalize=True,
    margins=True)
    )
```

```python
# exercise 3
# compare normalize=True vs normalize="index"
# what is the difference
# normalize=True shows the proportion of the grand total which is 2/12=0.1667
# normalize=True shows the proportion of the grand total (all 12 SKUs)
# normalize=True answers the question: what percent of all SKU's fall into this combination?
# What percentage of our total SKU count are East warehouse Widgets?
# Is inventory evenly distributed across the network?

# whereas normalize=Index shows the sum of that categories column which is 2/12 + 2/12 = 4/12
# normalize="index" shows the proportion of its row total, East has 6 SKU's total, and East/Component =2/6 = 0.33
# normalize="index" would answer: what percentage of East warehouse SKU's are components? 
# within each warehouse, what percentage are each category?


# normalize=True
print(pd.crosstab(index=df["warehouse"], 
                  columns=df["category"],
                  normalize=True)
                  )

# normalize="index"
print(pd.crosstab(index=df["warehouse"], 
                  columns=df["category"],
                  normalize="index")
                  )

# normalize = "columns"
# within each category, what percentage are in each warehouse?
# what percenage of components are in each warehouse?
print(pd.crosstab(index=df["warehouse"], 
                  columns=df["category"],
                  normalize="columns")
                  )

# normalize = True --> Grand Total in the denominator
# normalize = "index" --> Row Total in the denominator
# normalize = "columns" --> Column total in the denominator
```

```python
# exercise 4
# create a simple crosstab showing the total units on hand by warehouse and category
ct = pd.crosstab(
    index=df["warehouse"],
    columns=df["category"],
    values=df["units_on_hand"],
    aggfunc="sum"
)

print(ct)

# note that pd.crosstab() is more useful than .groupby()
# if you want a matrix layout with rows x columns without the need to reshape the data
# for frequency counts
# if you need normalization
# or quick exploratory analysis

# use .groupby().agg() when you need multiple aggregations at once
# named output columns or youre building a production level pipeline

# filter the warehouse where Widget units > 100
print(ct[ct["Widget"] > 100])
```
