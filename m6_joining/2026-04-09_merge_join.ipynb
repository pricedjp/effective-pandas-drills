```python
# M6.1 -- Joining dataframes with .merge() and .join()

# joining combines dataframes by matching rows on a shared key

# Basic Join Types
# Inner join  — keep only rows that match in BOTH DataFrames
# Left join   — keep ALL rows from the left, match where possible
# Right join  — keep ALL rows from the right, match where possible
# Outer join  — keep ALL rows from BOTH, match where possible

# import libraries
import pandas as pd
import numpy as np

# create two dataframes to join -- one for purchase orders and one for supplier masters
df_po = pd.DataFrame({
    "po_id":    ["PO-001", "PO-002", "PO-003", "PO-004", "PO-005"],
    "supplier": ["Acme", "GlobalCo", "Acme", "FastParts", "NewVendor"],
    "amount":   [1500, 2200, 800, 1200, 950],
    "category": ["Widget", "Gadget", "Component", "Widget", "Gadget"]
})

# supplier master
df_suppliers = pd.DataFrame({
    "supplier":    ["Acme", "GlobalCo", "FastParts"],
    "country":     ["USA", "UK", "Germany"],
    "rating":      [4.5, 3.8, 4.2],
    "approved":    [True, True, True]
})

print(df_po)
print(df_suppliers)
```

```python
# Exercise 2 - Inner Join
# Join PO's to supplier master on "supplier"
# Keep only PO's with a matching supplier

inner_join = df_po.merge(right=df_suppliers,
                         on="supplier",
                         how="inner")
print(inner_join)
```

```python
# Exercise 3 -- Left Join and Filter for Unegistered Suppliers
left_join = df_po.merge(right=df_suppliers,
                        on="supplier",
                        how="left",
                        indicator=True
                        )

print(left_join)

# filter for unregistered suppliers
# could chose any column that returned "NaN"
# np.nan() can not equal np.nan() in python you have to use pd.isna to detect NaN's
print(left_join[pd.isna(left_join["country"])])
```

```python
# Exercise 4 -- Outer Join -- Keep all rows from both dataframes

# when would you use each join type in sc procurement
# inner analyze POs from approved suppliers.
# left when you want ALL POs but need to flag which ones have unregistered suppliers.
# outer when you want to see both unregistered POs AND approved suppliers with no POs

outer_join = df_po.merge(right=df_suppliers,
                         on="supplier",
                         how="outer",
                         indicator=True
                         )
print(outer_join)
```
