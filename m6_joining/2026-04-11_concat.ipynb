```python
# M6.3 - concatenating dataframes with pd.concat()

# .merge() combines dataframes horizontally -- matching rows by a key column

# pd.concat() stacks dataframes either vertically (more rows) or horizontally (more columns)

# import libraries
import pandas as pd
import numpy as np

# create three monthly datasets
df_jan = pd.DataFrame({
    "po_id":    ["PO-001", "PO-002"],
    "supplier": ["Acme", "GlobalCo"],
    "amount":   [1500, 2200],
    "month":    ["Jan", "Jan"]
})

df_feb = pd.DataFrame({
    "po_id":    ["PO-003", "PO-004"],
    "supplier": ["FastParts", "Acme"],
    "amount":   [800, 1200],
    "month":    ["Feb", "Feb"]
})

df_mar = pd.DataFrame({
    "po_id":    ["PO-005", "PO-006"],
    "supplier": ["GlobalCo", "FastParts"],
    "amount":   [3100, 950],
    "month":    ["Mar", "Mar"]
})
```

```python
# Exercise 2

# Stack all 3 monthly dataframes into one
# axis=0 sets the index to zero-base
# the initial index would have duplicates like 0,1,0,1 so set ignore_index=True
df_all = pd.concat(objs=[df_jan, df_feb, df_mar], axis=0, ignore_index=True,)
print(df_all)
print(f"Total rows: {len(df_all)}")

# filter where amount > 1000
print(df_all[df_all["amount"] > 1000])
```

```python
# Exercise 3 
# use .concat() with keys and .xs()
df_keyed = pd.concat(
    [df_jan, df_feb, df_mar],
    keys=["January", "February", "March"] # labels dataframes in a hierarchical index where df_jan = "January"
)
print(df_keyed)

# select only the February rows using .xs()
# .xs(key, axis, level, drop_level) --> key is the label to select
# .xs() works on the index
print(df_keyed.xs("February"))
```
