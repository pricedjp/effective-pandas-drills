```python
# M7.3 -- Moving to Plotly

#import libraries
import pandas as pd
import numpy as np
import plotly.express as px

# create the dataframe
df = pd.DataFrame({
    "month":    ["Jan","Feb","Mar","Apr","May","Jun"] * 3,
    "supplier": ["Acme"]*6 + ["GlobalCo"]*6 + ["FastParts"]*6,
    "spend":    [1500,1800,1200,2100,1700,2000,
                 2200,1900,2400,2000,2300,2100,
                 800,950,1100,750,900,1050],
    "otd_rate": [80,90,70,100,90,100,
                 90,80,100,90,100,90,
                 70,80,90,70,80,90]
})

month_order = ["Jan","Feb","Mar","Apr","May","Jun"]
df["month"] = pd.Categorical(df["month"], categories=month_order, ordered=True)
df = df.sort_values("month")
print(df)
```

```python
# Exercie 2 -- Replicate M7.2 Charts Using Plotly

# multi-line chart with one line per supplier
fig = px.line(
    df,
    x="month",
    y="spend",
    color="supplier",
    markers=True,
    title="Monthly Spend by Supplier"
)
fig.show()

# grouped bar chart
fig = px.bar(
    df,
    x="month",
    y="spend",
    color="supplier",
    barmode="group",
    title="Monthly Spend by Supplier -- Grouped"
)
fig.show()
```

```python
# Exercise 3 -- Plotly Only Features
fig = px.scatter(
    df,
    x="spend",
    y="otd_rate",
    color="supplier",
    trendline="ols",
    title="Spend vs OTD Rate with Trendline"
)
fig.show()

# create a facet chart
# each supplier is a facet -- generating its own subplot
fig = px.bar(
    df,
    x="month",
    y="spend",
    facet_col="supplier",
    title="Monthly Spend -- Faceted by Supplier"
)
fig.show()
```
